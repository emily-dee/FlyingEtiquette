```python
flying_df['Height (Inches)'] = flying_df['How tall are you?'].replace({
        '6\'6" and above': '6\'6"',
        'Under 5 ft.': '5\'0"'
})
```

```python
flying_df['Height (Inches)'].unique()
```

## Converting values to inches with function

```python
def height_to_inches(tall: str) -> int:
    if pd.isna(tall):
        return tall
    feet = int(tall.split("'")[0])
    leftover_inches = int(tall.split("'")[1].split('"')[0])
    inches = (feet * 12) + leftover_inches
    return inches
```

```python
flying_df['Height (Inches)'] = flying_df['Height (Inches)'].apply(height_to_inches)
```

```python
get_up_q = "On a 6 hour flight from NYC to LA, how many times is it acceptable to get up if you're not in an aisle seat?"
flying_df["On a 6 hour flight from NYC to LA, how many times is it acceptable to get up if you're not in an aisle seat?"].unique()
```

```python
flying_df['Get Up Count'] = flying_df["On a 6 hour flight from NYC to LA, how many times is it acceptable to get up if you're not in an aisle seat?"]
flying_df['Get Up Count']
```

```python
def get_up_count_convert(answer: str) -> int:
    if answer == 'It is not okay to get up during flight':
        return 0
    elif answer == 'Once':
        return 1
    elif answer == 'Twice':
        return 2
    elif answer == 'Three times':
        return 3
    elif answer == 'Four times':
        return 4
    elif answer == 'More than five times times':
        return 5
    else:
        return answer
```

```python
flying_df['Get Up Count'] = flying_df['Get Up Count'].apply(get_up_count_convert)
flying_df['Get Up Count']
```

I left the NaN values as null since the value is unknown, and I gave "More than five times" a value of 5, I feel that shows an accurate bell curve that would naturally drop off shortly after 5

```python
flying_df['Get Up Count'].unique()
```

```python
sns.countplot(data=flying_df, x='Get Up Count',
            hue='Age', hue_order=['18-29', '30-44', '45-60', '> 60']).set(title='How many times is it appropriate to get up if you are not in the aisle seat?', xlabel='Survey Response')
```

```python
flying_df.groupby('Age')['Get Up Count'].mean()
```

45-60 has the highest average number but all are close to 2.8, which aligns with count plot above

Converting the How often column to be a rough yearly estimate

```python
def year_count_convert(answer: str) -> int:
    if answer == 'Once a year or less':
        return 1
    elif answer == 'Once a month or less':
        return 12
    elif answer == 'Never':
        return 0
    elif answer == 'A few times per week':
        return 156
    elif answer == 'Every day':
        return 365
    elif answer == 'A few times per month':
        return 36
    else:
        return answer
```

```python
flying_df['Yearly Flight Count'] = flying_df['How often do you travel by plane?'].apply(year_count_convert)
flying_df['Yearly Flight Count'].unique()
```

```python
flying_df.groupby('Household Income')['Yearly Flight Count'].mean()
```
